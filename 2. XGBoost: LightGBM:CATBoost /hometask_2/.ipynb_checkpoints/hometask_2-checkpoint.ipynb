**Задание 2:** обработать категориальные признаки любым способом (который вы знаете) и добавить их к данным. Выполнить задание 1.

```python
categorical_features_train = train_data.select_dtypes(include=[np.object])
categorical_features_test = test_data.select_dtypes(include=[np.object])
print(f"Categorical Feature Count {categorical_features_train.shape[1]}")
categorical_features_train.head(n=10)
```

```python
missing_values = categorical_features_train.isnull().sum(axis=0).reset_index()
missing_values.columns = ['column_name', 'missing_count']
missing_values = missing_values.loc[missing_values['missing_count']>0]
missing_values = missing_values.sort_values(by='missing_count')

ind = np.arange(missing_values.shape[0])
fig, axes = plt.subplots(figsize=(15, 8))
rects = axes.barh(ind, missing_values.missing_count.values, height=0.9)
axes.set_yticks(ind)
axes.set_yticklabels(missing_values.column_name.values, rotation='horizontal')
axes.set_title("Missing Observations Count - Categorical Features")
axes.set_xlabel("Missing Observations Count")
```

```python
#Nullity correlation ranges from -1 (if one variable appears the other definitely does not) to 
#0 (variables appearing or not appearing have no effect on one another) to 1 
#(if one variable appears the other definitely also does).

msno.heatmap(categorical_features_train)
```

```python
#leave only features when there are less than 50% of values are missing
categorical_names = [name for name, missing in categorical_features_train.isnull().sum(axis=0).iteritems() if missing < train_data.shape[0]/2]
categorical_names
```

```python
categorical_features_train = categorical_features_train[categorical_names]
categorical_features_test = categorical_features_test[categorical_names]
```

```python
for column in categorical_names:
    categorical_features_train[column].fillna(categorical_features_train[column].mode()[0], inplace=True)
    categorical_features_test[column].fillna(categorical_features_test[column].mode()[0], inplace=True)
```

```python
categorical_features_train
```

```python
categorical_features_train_encoded = categorical_features_train.copy()
categorical_features_test_encoded = categorical_features_test.copy()
```

```python
le = LabelEncoder() 

for column in categorical_names:    
    categorical_features_train_encoded[column]= le.fit_transform(categorical_features_train_encoded[column])
    categorical_features_test_encoded[column]= le.transform(categorical_features_test_encoded[column])   
    
categorical_features_train_encoded.head(2)
```

```python
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(categorical_features_train_encoded[categorical_names]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(categorical_features_test_encoded[categorical_names]))

# Adding column names to the encoded data set.
OH_cols_train.columns = OH_encoder.get_feature_names(categorical_names)
OH_cols_valid.columns = OH_encoder.get_feature_names(categorical_names)

# One-hot encoding removed index; put it back
OH_cols_train.index = categorical_features_train_encoded.index
OH_cols_valid.index = categorical_features_test_encoded.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = categorical_features_train_encoded.drop(categorical_names, axis=1)
num_X_valid = categorical_features_test_encoded.drop(categorical_names, axis=1)

# Add one-hot encoded columns to numerical features
categorical_features_train_encoded = pd.concat([num_X_train, OH_cols_train], axis=1)
categorical_features_test_encoded = pd.concat([num_X_valid, OH_cols_valid], axis=1)
```

```python
categorical_features_train_encoded.head(2)
```

```python
train_data_full = pd.concat([categorical_features_train_encoded, train_data_numeric], axis=1)
test_data_full = pd.concat([categorical_features_test_encoded, test_data_numeric], axis=1)
```

```python
train_data_full.head(2)
```

```python
dtrain_categoric = xgb.DMatrix(
    data=train_data_full, label=train_target
)
dtest_categoric = xgb.DMatrix(
    data=test_data_full, label=test_target
)
```

```python
params = {
    "booster": "gbtree",
    "objective": "binary:logistic",
    "eval_metric": "auc",
    "learning_rate": 0.1,
    "n_estimators": 1000,
    "reg_lambda": 100,
    "max_depth": 4,
    "gamma": 10,
    "nthread": 6,
    "seed": 27
}
```

```python
model_with_categoric = xgb.train(
    params=params,
    dtrain=dtrain_categoric,
    num_boost_round=1000,
    early_stopping_rounds=100,
    evals=[(dtrain_categoric, "train"), (dtest_categoric, "test")],
    verbose_eval=20,
    maximize=True,
)
```

**Задание 10:** построить ROC-кривую для всех построенных алгоритмов на обучающей и тестовой выборке. Сделать выводы о работе алгоритмов с точки зрения качества на тестовой выборке и с точки зрения переобучения.

```python
#####test######
CAT_with_categoric_by_default_test = model_cat_with_categoric_by_default.predict_proba(test_data_full)[:,1]
fpr, tpr, thresholds = roc_curve(test_target, CAT_with_categoric_by_default_test)
roc_auc = roc_auc_score(test_target, CAT_with_categoric_by_default_test)
print("CAT_with_categoric_by_default_test:", round(roc_auc,2))

XGBoost_with_categoric_test = model_with_categoric.predict(dtest_categoric)
fpr2, tpr2, thresholds2 = roc_curve(test_target, XGBoost_with_categoric_test)
roc_auc2 = roc_auc_score(test_target, XGBoost_with_categoric_test)
print("XGBoost_with_categoric_test:", round(roc_auc2,2))

LGBM_with_categoric_by_default_test = model_lgb_with_categoric_by_default.predict(test_data_full)
fpr1, tpr1, thresholds1 = roc_curve(test_target, LGBM_with_categoric_by_default_test)
roc_auc1 = roc_auc_score(test_target, LGBM_with_categoric_by_default_test)
print("LGBM_with_categoric_by_default_test:", round(roc_auc1,2))
#####train######
CAT_with_categoric_by_default_train = model_cat_with_categoric_by_default.predict_proba(train_data_full)[:,1]
fpr3, tpr3, thresholds3 = roc_curve(train_target, CAT_with_categoric_by_default_train)
roc_auc3 = roc_auc_score(train_target, CAT_with_categoric_by_default_train)
print("CAT_with_categoric_by_default_train:", round(roc_auc3,2))

XGBoost_with_categoric_train = model_with_categoric.predict(dtrain_categoric)
fpr4, tpr4, thresholds4 = roc_curve(train_target, XGBoost_with_categoric_train)
roc_auc4 = roc_auc_score(train_target, XGBoost_with_categoric_train)
print("XGBoost_with_categoric_train:", round(roc_auc4,2))

LGBM_with_categoric_by_default_train = model_lgb_with_categoric_by_default.predict(train_data_full)
fpr5, tpr5, thresholds5 = roc_curve(train_target, LGBM_with_categoric_by_default_train)
roc_auc5 = roc_auc_score(train_target, LGBM_with_categoric_by_default_train)
print("LGBM_with_categoric_by_default_train:", round(roc_auc5,2))

plt.figure(figsize=(25,15))
plt.plot(fpr, tpr, label='CAT_test (area = %0.3f)' % roc_auc)
plt.plot(fpr1, tpr1, label='LGBM_test (area = %0.3f)' % roc_auc1)
plt.plot(fpr2, tpr2, label='XG_test (area = %0.3f)' % roc_auc2)

plt.plot(fpr3, tpr3, label='CAT_train (area = %0.3f)' % roc_auc3)
plt.plot(fpr5, tpr5, label='LGBM_train (area = %0.3f)' % roc_auc5)
plt.plot(fpr4, tpr4, label='XG_train (area = %0.3f)' % roc_auc4)


plt.legend()
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
